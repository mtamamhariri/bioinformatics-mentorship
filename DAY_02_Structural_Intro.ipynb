```python
# Kode yang disempurnakan: Tampilkan protein sebagai kartun dan ligan sebagai "ball+stick"
view = nv.show_file(pdb_file, ext="pdb")

# Hapus representasi 'kartun' default
view.clear_representations()

# 1. Tampilkan protein (polimer) sebagai kartun
view.add_representation('cartoon', selection='protein')

# 2. Tampilkan ligan (bukan air & bukan polimer) sebagai 'ball+stick'
view.add_representation('ball+stick', selection='not protein and not water')

# Tampilkan visualisasi
view
```
